```python
tips.drop(columns = tips.loc[:, tips.columns.str.contains('^Unnamed')], inplace = True)
tips
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
yes_tips = tips[tips.smoker == 'Yes']
no_tips = tips[tips.smoker == 'No']
size, (ax1,ax2) = plt.subplots(2, figsize = (10,7))
ax1.scatter(x = yes_tips.total_bill  , y = yes_tips.tip, c = np.where(yes_tips.sex == 'Male','r','y'));
ax1.title.set_text('smoker')
ax2.scatter(x = no_tips.total_bill, y = no_tips.tip, c = np.where(no_tips.sex == 'Male','r','y'));
ax2.title.set_text('not smoker')
```
