```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
fire = pd.read_csv("fire_nrt_V1_96617.csv")
```

```python
pre = {}
for i in fire.confidence.unique().tolist():
    pre[i] = fire[fire.confidence == i]['bright_ti4'].mean()
```

```python
pre
```

```python
# sns.set_style("darkgrid")
figure, axis = plt.subplots(1,2, figsize = (10,5))

axis[0].set_xlabel("latitude")
axis[0].set_ylabel("longitude")
axis[0].hexbin(fire.latitude, fire.longitude, C = fire.scan, cmap ='Greens', alpha = 1, gridsize = 75);

axis[1].set_xlabel("latitude")
axis[1].set_ylabel("longitude")
axis[1].scatter(fire.latitude, fire.longitude,  alpha = 1, s = 2, color = np.where(fire.confidence == 'n','g','r'));
```
